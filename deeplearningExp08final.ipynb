```python
# --------------------------------------------------------------
# SENTIMENT ANALYSIS USING SIMPLE RNN (ONE-CELL CODE FOR COLAB)
# --------------------------------------------------------------

import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# ------------------------- Step 1: Load Dataset -------------------------
vocab_size = 10000
(max_length, embedding_dim) = (200, 64)

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)

# ------------------------- Step 2: Preprocess ---------------------------
X_train = pad_sequences(X_train, maxlen=max_length)
X_test = pad_sequences(X_test, maxlen=max_length)

# ------------------------- Step 3: Build Model --------------------------
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    SimpleRNN(64),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# ------------------------- Step 4: Train Model --------------------------
history = model.fit(X_train, y_train, epochs=3, batch_size=128, validation_split=0.2)

# ------------------------- Step 5: Evaluate -----------------------------
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\nTest Accuracy: {accuracy:.4f}")

# ------------------------- Step 6: Predict on Custom Input --------------
word_index = imdb.get_word_index()

def encode_text(text):
    tokens = text.lower().split()
    encoded = [word_index.get(word, 2) for word in tokens]  # 2 = <UNK>
    return pad_sequences([encoded], maxlen=max_length)

user_review = "The movie was amazing and I enjoyed every moment!"
encoded = encode_text(user_review)
prediction = model.predict(encoded)[0][0]

print("\nReview:", user_review)
print("Prediction Score:", prediction)
print("Sentiment:", "ğŸ‘ Positive" if prediction >= 0.5 else "ğŸ‘ Negative")
```
