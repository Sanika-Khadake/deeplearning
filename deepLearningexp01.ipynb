# Step 1: Import TensorFlow v1 compatibility mode

```python
import tensorflow as tf
#The compat.v1 module allows you to write or run code using the TensorFlow 1.x style in TensorFlow 2.x.

```

```python
tf.compat.v1.disable_v2_behavior()

#This line disables TensorFlow 2.x features, such as:
#Eager execution
#Updated APIs
#It essentially makes the environment behave like TensorFlow 1.x.
#You need this to run 1.x-style code (like using sessions and graphs).
```


# Step 2: Define constants

```python
input1 = tf.constant(1.8, name="input1")
input2 = tf.constant(3.6, name="input2")
```

# Step 3: Apply round and floor operations

```python
rounded = tf.round(input1, name="round")     # round(1.8) -> 2.0
floored = tf.floor(input2, name="floor")     # floor(3.6) -> 3.0
```

# Step 4: Addition node (round + floor)

```python
add_left = tf.add(rounded, floored, name="add_left")  # 2.0 + 3.0 = 5.0
```

# Step 5: abs node (abs(add_left))

```python
abs_node = tf.abs(add_left, name="abs")  # abs(5.0) = 5.0
```

# Step 6: Multiply rounded and add_left

```python
multiply = tf.multiply(rounded, add_left, name="multiply")  # 2.0 * 5.0 = 10.0
```

# Step 7: Final addition node (multiply + abs)

```python
final_result = tf.add(multiply, abs_node, name="final_add")  # 10.0 + 5.0 = 15.0
```

# Step 8: Execute with a session

```python
with tf.compat.v1.Session() as sess:
    result = sess.run(final_result)
    print("Final Result =", result)

#This starts a TensorFlow session, which is used to run the computation graph.
#The with block ensures the session is properly closed after use.
#sess.run(final_result)
#This executes the graph up to the final_result node.
#It returns the evaluated result as a NumPy value.
```
